__This starter notebook uses the [DistilBERT](https://arxiv.org/abs/1910.01108) pretrained model from KerasNLP.__


**BERT** stands for **Bidirectional Encoder Representations from Transformers**. BERT and other Transformer encoder architectures have been wildly successful on a variety of tasks in NLP (natural language processing). They compute vector-space representations of natural language that are suitable for use in deep learning models.

The BERT family of models uses the **Transformer encoder architecture** to process each token of input text in the full context of all tokens before and after, hence the name: Bidirectional Encoder Representations from Transformers.

BERT models are usually pre-trained on a large corpus of text, then fine-tuned for specific tasks.

**DistilBERT model** is a distilled form of the **BERT** model. The size of a BERT model was reduced by 40% via knowledge distillation during the pre-training phase while retaining 97% of its language understanding abilities and being 60% faster.

# Create the model

```python
# We choose 512 because it's the limit of DistilBert
SEQ_LENGTH = 512

# Use a shorter sequence length.
preprocessor = keras_nlp.models.DistilBertPreprocessor.from_preset(
    "distil_bert_base_en_uncased",
    sequence_length=SEQ_LENGTH,
)

# Pretrained classifier.
classifier = keras_nlp.models.DistilBertClassifier.from_preset(
    "distil_bert_base_en_uncased",
    num_classes=2,
    activation=None,
    preprocessor=preprocessor,
)

# Re-compile (e.g., with a new learning rate)
classifier.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(5e-4),
    metrics=[
        keras.metrics.SparseCategoricalAccuracy()
   ]
)
    

# Access backbone programmatically (e.g., to change `trainable`).
classifier.backbone.trainable = False


classifier.summary()
```

```python
# Split the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train_essays_final["text"],
                                                    df_train_essays_final["generated"],
                                                    test_size=0.33,
                                                    random_state=42)
```

```python
# Fit
classifier.fit(x=X_train, 
               y=y_train,
               validation_data=(X_test, y_test),
               epochs=1,
               batch_size=64
              )
```

```python
def displayConfusionMatrix(y_true, y_pred, dataset):
    disp = ConfusionMatrixDisplay.from_predictions(
        y_true,
        np.argmax(y_pred, axis=1),
        display_labels=["Not Generated","Generated"],
        cmap=plt.cm.Blues
    )

    tn, fp, fn, tp = confusion_matrix(y_true, np.argmax(y_pred, axis=1)).ravel()
    f1_score = tp / (tp+((fn+fp)/2))

    disp.ax_.set_title("Confusion Matrix on " + dataset + " Dataset -- F1 Score: " + str(f1_score.round(2)))
```

```python
y_pred_test = classifier.predict(X_test)
```

```python
import pandas as pd
from nltk import FreqDist
from nltk.tokenize import word_tokenize

# Assuming 'df_train_essays_final' is your DataFrame and it includes 'text' for the essay texts
# and 'generated' as a binary indicator (1 for AI-generated, 0 for human-written)

# Filter the DataFrame for AI-generated and human-written texts
ai_texts = df_train_essays_final[df_train_essays_final['generated'] == 1]['text']
human_texts = df_train_essays_final[df_train_essays_final['generated'] == 0]['text']

# Tokenize and compute frequency distribution for each category
ai_words = [word.lower() for text in ai_texts for word in word_tokenize(text)]
human_words = [word.lower() for text in human_texts for word in word_tokenize(text)]

ai_freq = FreqDist(ai_words)
human_freq = FreqDist(human_words)

# Normalize frequencies by total number of words
total_ai_words = sum(ai_freq.values())
total_human_words = sum(human_freq.values())
ai_freq_norm = {word: freq / total_ai_words for word, freq in ai_freq.items()}
human_freq_norm = {word: freq / total_human_words for word, freq in human_freq.items()}

# Calculate the difference in frequencies
diff_freq = {word: ai_freq_norm.get(word, 0) - human_freq_norm.get(word, 0) for word in set(ai_freq_norm) | set(human_freq_norm)}

# Filter to get words more frequent in AI-generated texts and sort them
more_common_in_ai = {word: freq for word, freq in diff_freq.items() if freq > 0}
sorted_more_common_in_ai = sorted(more_common_in_ai.items(), key=lambda item: item[1], reverse=True)

# Display the top words
print(sorted_more_common_in_ai[:20])  # Adjust the slice for more or fewer words
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk import FreqDist, word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud

# Download NLTK stop words if not already downloaded
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Load stop words
stop_words = set(stopwords.words('english'))

# Assuming 'df_train_essays_final' is your DataFrame
ai_texts = df_train_essays_final[df_train_essays_final['generated'] == 1]['text']
human_texts = df_train_essays_final[df_train_essays_final['generated'] == 0]['text']

# Tokenize and compute frequency distribution for each category
ai_words = [word.lower() for text in ai_texts for word in word_tokenize(text) if word.lower() not in stop_words and word.isalpha()]
human_words = [word.lower() for text in human_texts for word in word_tokenize(text) if word.lower() not in stop_words and word.isalpha()]

ai_freq = FreqDist(ai_words)
human_freq = FreqDist(human_words)

# Normalize frequencies by total number of words
total_ai_words = sum(ai_freq.values())
total_human_words = sum(human_freq.values())
ai_freq_norm = {word: freq / total_ai_words for word, freq in ai_freq.items()}
human_freq_norm = {word: freq / total_human_words for word, freq in human_freq.items()}

# Calculate the difference in frequencies and filter out common words
diff_freq = {word: ai_freq_norm.get(word, 0) - human_freq_norm.get(word, 0) 
             for word in set(ai_freq_norm) | set(human_freq_norm) 
             if ai_freq_norm.get(word, 0) > 0 and human_freq_norm.get(word, 0) == 0}

# Create a word cloud from the differential frequencies
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(diff_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
